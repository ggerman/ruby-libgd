ruby-libgd Gradient

```ruby
require "gd"
# this function show the image on the notebook
def render(file_name)
    IRuby.display "<img src='./#{file_name}'>", mime: "text/html"
end
# calculate step between lines to generate the gradient
def step source, target, steps
    source.zip(target).map { |s, t| (t - s).to_f / (steps - 1)}
end
```

```ruby
width = 400
steps = height = 400

image = GD::Image.new(width, height)

source_rgb = [0, 48, 64]
target_rgb = [54, 54, 54]

step = step(source_rgb, target_rgb, height)

steps.times do |i|
    rgb = source_rgb.zip(step).map { |s, st| (s + st * i).round }
    current_color = GD::Color.rgb(rgb[0], rgb[1], rgb[2])
    image.line(0, i, width-1, i, current_color)
end

```

```ruby
def random_font
    # system font path
    path = "/usr/share/fonts/truetype/**/*.ttf"
    Dir.glob(path).sample
end
```

```ruby
date = Time.now
font = random_font
font_color = GD::Color.rgb(250, 250, 250)
size = 20

string = "Gradient"

w, h = image.text_bbox(string, :font => font, :size => size)

image.text(
    string,
    x: (width / 2) - (w / 2), 
    y: (height / 2)- (h / 2), 
    color: font_color, 
    size: 20, 
    font: font
)

file_name = "colors.jpeg"
image.save("/work/#{file_name}")

render(file_name)
```
