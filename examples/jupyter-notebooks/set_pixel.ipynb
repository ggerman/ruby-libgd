```ruby
require 'gd'

# this function show the image on the notebook
def render(file_name)
    IRuby.display "<img src='#{file_name}' />", mime: "text/html"
end
```

```ruby
width = 250
height = 250

img = GD::Image.new(width, height)

width.times do |i|
    height.times do |j|
        img.set_pixel(i, j, [rand(250), rand(250), rand(250)])
    end
end

name = "random_pixel.png"
img.save("/work/#{name}")
render(name)
```
